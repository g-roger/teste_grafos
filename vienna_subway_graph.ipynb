```python
import networkx as nx 
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vienna_subway.csv', delimiter=';')
```

```python
df.head()
```

```python
df.info()
```

```python
vienna_sub = nx.from_pandas_edgelist(df, source='Start', target='Stop')
```

```python
plt.figure(figsize=(30,20))
nx.draw_networkx(vienna_sub)
plt.show()
```

```python
nx.info(vienna_sub)
```

```python
nx.degree_centrality(vienna_sub)
```

```python
nx.shortest_path(vienna_sub, "Floridsdorf", "Landstrasse")
```

```python
vienna_sub.edges
```

```python
# Grus de conexões
nx.degree(vienna_sub)
```

```python
# quantas estações conectadas com Thaliastresse
vienna_sub.degree('Thaliastrasse')
```

```python
vienna_sub.neighbors("Oberlaa")
```

```python
nx.all_neighbors(vienna_sub,"Oberlaa")
```

```python
new_group = nx.bfs_tree(vienna_sub, "Krieau")
```

```python
new_group.nodes
```
